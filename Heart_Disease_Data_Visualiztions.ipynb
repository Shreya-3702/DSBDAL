Attribute information :

1. **age**: Age in years
2. **sex**: Gender (1 = male; 0 = female)
3. **cp**: Chest pain type
   - Value 1: Typical angina
   - Value 2: Atypical angina
   - Value 3: Non-anginal pain
   - Value 4: Asymptomatic
4. **trestbps**: Resting blood pressure (in mm Hg on admission to the hospital)
5. **chol**: Serum cholesterol in mg/dl
6. **fbs**: Fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
7. **restecg**: Resting electrocardiographic results
   - Value 0: Normal
   - Value 1: Having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
   - Value 2: Showing probable or definite left ventricular hypertrophy by Estes' criteria
8. **thalach**: Maximum heart rate achieved
9. **exang**: Exercise induced angina (1 = yes; 0 = no)
10. **oldpeak**: ST depression induced by exercise relative to rest
11. **slope**: The slope of the peak exercise ST segment
    - Value 1: Upsloping
    - Value 2: Flat
    - Value 3: Downsloping
12. **ca**: Number of major vessels (0-3) colored by fluoroscopy
13. **thal**: Thalassemia
    - Value 3: Normal
    - Value 6: Fixed defect
    - Value 7: Reversible defect
14. **num**:(the predicted attribute)
      Diagnosis of heart disease (angiographic disease status)
    - Value 0: < 50% diameter narrowing
    - Value 1: > 50% diameter narrowing
      (in any major vessel: attributes 59 through 68 are vessels)



```python
import pandas as pd
```

```python
df = pd.read_csv('Cleavland.csv',header=None)
```

```python
column_names = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']
df.columns = column_names
```

```python
df
```

```python
(df== '?').sum()
```

```python
import numpy as np
df.replace('?',np.nan)
```

```python
df.info()
```

```python
columns_to_convert = ['ca', 'thal']

for col in columns_to_convert:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.info()
```

```python
df.isna().sum()
```

```python
df = df.fillna(df.median())
df.isna().sum()
```

# Data Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**Histogram**

**To visualize the distribution of a single variable, such as age, cholesterol levels, etc., to understand its frequency distribution.**

```python
plt.hist(df['age'],edgecolor = 'black')
plt.title('Histogram For Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.hist(df['chol'],edgecolor = 'black')
plt.title('Histogram For Distribution of cholestrol levels')
plt.xlabel('Chol')
plt.ylabel('Frequency')
plt.show()
```

**Dot Plot**

**To compare two variables, such as age and maximum heart rate achieved, by plotting individual data points along a single axis**

```python
plt.figure(figsize=(20,9))
sns.stripplot(x='age', y='thalach', data=df)
plt.xlabel('Age')
plt.ylabel('Maximum Heart Rate Achieved')
plt.title('Dot Plot of Age vs Maximum Heart Rate Achieved')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.stripplot(x='cp', y='trestbps', data=df, jitter=True)
plt.xlabel('Chest Pain Type')
plt.ylabel('Resting Blood Pressure')
plt.title('Chest Pain vs. Resting Blood Pressure')
plt.show()
```

**Bar Plot**

**c. Bar Plots: Objective - To compare the frequency of different categories within a categorical variable (e.g., chest pain type, sex,restgc,thal etc).**

```python
count = df['cp'].value_counts()
plt.bar(count.index,count)
plt.xlabel('Chest Pain Type')
plt.ylabel('Frequency')
plt.title('Bar Plot of Chest Pain Type')
plt.show()
```

```python
count = df['sex'].value_counts()
plt.bar(count.index,count)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Bar Plot of Gender')
plt.show()
```

**Line Charts**

**To visualize trends in a continuous variable over time (e.g., thalach).**



```python

plt.plot(df['thalach'], linestyle='-')
plt.xlabel('index')
plt.ylabel('Heart Rate')
plt.title('Line Chart of Heart Rate over Time')
plt.show()
```

**e. Add Histogram and Scatter plot to box plot.**

```python
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.hist(df['age'],edgecolor = 'black')
plt.title('Histogram For Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')


plt.subplot(2,2,2)
sns.scatterplot(df,x = 'age',y = 'thalach',hue="num")
plt.xlabel('Age')
plt.ylabel('Heart Rate')
plt.title('Age vs. Heart Rate')


plt.subplot(2,2,3)
sns.boxplot(df)
plt.title('Box plot')
plt.xlabel('Features')
plt.ylabel('Values')

plt.tight_layout()
plt.show()

```

f.**Pie Charts**

**To show the proportion of different categories within a categorical variable**

```python
count = df['sex'].value_counts()
plt.pie(count,labels=['M', 'F'],colors=['skyblue', 'lightgreen'],autopct='%1.2f%%')
plt.title('Male Female Proportion  in %')
plt.show()
```

```python
count = df['cp'].value_counts()
plt.pie(count,labels = ['Typical angina','Atypical angina','Non-anginal pain','Asymptomatic'],autopct='%1.2f%%')
plt.title('Chest Pain Type Categories Proportion in %')
plt.show()
```

```python
count = df['thal'].value_counts()
plt.pie(count,labels = ['Normal','Fixed defect','Reversible defect'],autopct='%1.2f%%')
plt.title('Thalassemia Proportion in %')
plt.show()
```

**g. Box Plot**

```python
plt.figure(figsize=(12,8))
sns.boxplot(df)
plt.title('Box plot')
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```
