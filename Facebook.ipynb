**a. Create data subsets for type of post**

```python
df['Type'].unique()
```

```python
X = df[df['Type']=='Photo']
X
```

```python
subset1 = X.iloc[:10,:7]
subset1
```

```python
subset2 = X.iloc[10:20,:7]
subset2
```

**b. Merge two subsets**

```python
pd.concat([subset1,subset2])
```

**c. Sort Data on Page total likes**

```python
df.sort_values('Page total likes')
```

```python
df.sort_values('Page total likes', ascending=False)
```

**d. Transposing Data**

```python
df.transpose()
```

**e. Melting Data to long format**

```python
Y = df.melt(id_vars=['Type'],var_name = 'PostInfo')

Y['PostInfo'].unique()
```

```python
Y
```

```python
Z = df.melt(id_vars=['Type'],var_name='PostInfo',value_vars=['like','share','comment'])
Z
```

```python
#grouped_data = Z.groupby(['Type', 'PostInfo'])['value'].sum()
#grouped_data
Z.pivot_table(index=['Type','PostInfo'], values='value') #default aggfunc is mean
```

**f. Casting data to wide format**



```python
df.pivot_table(index=['Type','Category'], values='Page total likes',aggfunc='sum')
```
